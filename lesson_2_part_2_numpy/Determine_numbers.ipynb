```python
#All possible
possibles = []
for i in range(2, 100):
    for j in range(i, 100):
        possibles.append([i,j,i+j,i*j])
```

```python
len(possibles)
```

```python
#Prod: I don't know the numbers

product_count = {}

for item in possibles:
    prod = item[3]
    if prod in product_count:
        product_count[prod] += 1
    else:
        product_count[prod] = 1
        
filtered = []
for item in possibles:
    if product_count[item[3]] > 1:
        filtered.append(item)

        
```

```python
#Sum: I know it

correct_sums = {}

for item in possibles:
    suma = item[2]
    if suma in correct_sums:
        if correct_sums[suma] == 0:
            continue
    if product_count[item[3]] > 1:
        correct_sums[suma] = 1
    else:
        correct_sums[suma] = 0

filtered2 = []
for item in possibles:
    if correct_sums[item[2]] == 1:
        filtered2.append(item)
```

```python
filtered2
```

```python
#Prod: then I know the numbers

product_count = {}

for item in filtered2:
    if item[3] in product_count:
        product_count[item[3]] += 1
    else:
        product_count[item[3]] = 1

filtered3 = []
for item in filtered2:
    if product_count[item[3]] == 1:
        filtered3.append(item)
```

```python
filtered3
```

```python
#Sum: then I know the numbers too.

sum_count = {}

for item in filtered3:
    if item[2] in sum_count:
        sum_count[item[2]] += 1
    else:
        sum_count[item[2]] = 1
        
filtered4 = []
for item in filtered3:
    if sum_count[item[2]] == 1:
        filtered4.append(item)
```
